```python
from PIL import Image

def encrypt_image(image_path, key):
    # Open the image
    image = Image.open(image_path)
    pixels = image.load()

    # Apply XOR operation to each pixel
    for i in range(image.width):
        for j in range(image.height):
            r, g, b = pixels[i, j]
            pixels[i, j] = (r ^ key, g ^ key, b ^ key)

    # Save the encrypted image
    encrypted_path = 'encrypted_image.png'
    image.save(encrypted_path)

    return f'Image encrypted successfully. Encrypted image saved as: {encrypted_path}'

def decrypt_image(encrypted_image_path, key):
    # Open the encrypted image
    image = Image.open(encrypted_image_path)
    pixels = image.load()

    # Apply XOR operation to each pixel to decrypt
    for i in range(image.width):
        for j in range(image.height):
            r, g, b = pixels[i, j]
            pixels[i, j] = (r ^ key, g ^ key, b ^ key)

    # Save the decrypted image
    decrypted_path = 'decrypted_image.png'
    image.save(decrypted_path)

    return f'Image decrypted successfully. Decrypted image saved as: {decrypted_path}'

# Example usage
image_path = '/content/kingfisher-fishing_2023-03-04-235438_oomf.jpg'
encryption_key = 42

# Encrypt the image
encryption_result = encrypt_image(image_path, encryption_key)
print(encryption_result)

# Decrypt the image
decryption_result = decrypt_image('encrypted_image.png', encryption_key)
print(decryption_result)
```
